```python
def COVID_19_REPORTED_OCCURENCES(VAX_TYPE):
    if 'COVID19' in VAX_TYPE:
        return True
    else:
        return False
COVID19_REPORTED = sum(df3['VAX_TYPE'].apply(lambda x: COVID_19_REPORTED_OCCURENCES(x)))
print(f"Total Reported VAERS 2021 occurences for COVID19 vaccines: {COVID19_REPORTED}")
```

```python
def COVID_19_REPORTED_OCCURENCES_PFIZER(VAX_NAME):
    if 'PFIZER' in VAX_NAME:
        return True
    else:
        return False
    
# Sum of the total Reported occurences of pfizer
PFIZER_REPORTED = df3['VAX_NAME'].apply(lambda x: COVID_19_REPORTED_OCCURENCES_PFIZER(x))
PFIZER_REPORTED_SUM = sum(PFIZER_REPORTED)
print(f"Total Reported VAERS 2021 occurences for Pfizer vaccine: {PFIZER_REPORTED_SUM}")

# Gathering all the VAERS_IDs for Pfizer
filtered_VAERSID_Pfizer = df3[PFIZER_REPORTED]['VAERS_ID']
print (f"The VAERSIDs reported for pFizer are stored in filtered_VAERSID_Pfizer \n {filtered_VAERSID_Pfizer}")
```

```python
def COVID_19_REPORTED_OCCURENCES_MODERNA(VAX_NAME):
    if 'MODERNA' in VAX_NAME:
        return True
    else:
        return False
MODERNA_REPORTED = df3['VAX_NAME'].apply(lambda x: COVID_19_REPORTED_OCCURENCES_MODERNA(x))
MODERNA_REPORTED_SUM = sum(MODERNA_REPORTED)
print(f"Total Reported VAERS 2021 occurences for pfizer Moderna: {MODERNA_REPORTED_SUM}")

# Gathering all the VAERS_IDs for Moderna
filtered_VAERSID_Moderna = df3[MODERNA_REPORTED]['VAERS_ID']
print (f"The VAERSIDs reported for pFizer are stored in filtered_VAERSID_Pfizer \n {filtered_VAERSID_Moderna}")
```

```python
# Counting unique Labels under VAX_NAME
label_counts = df3['VAX_NAME'].value_counts()
```

```python
plt.figure(figsize=(25, 10))
plt.bar(label_counts.index, label_counts.values)
plt.xlabel('Labels')
plt.ylabel('Number of Occurrences')
plt.title('Occurrences of all Unique Labels')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.bar(label_counts.index[:5], label_counts.values[:5])
plt.xlabel('Vaccine Type')
plt.ylabel('Number of Reported VAERS cases for each vaccine Type')
plt.title('Top 5 reported for Adverse effects based on Vaccine Type in 2021')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
